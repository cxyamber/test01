```python
import requests
def getNames(num):
    times = 1
    cnames = []
    for x in range(times):
        url = 'http://www.richyli.com/name/index.asp'
        r = requests.get(url)
        r.encoding = 'big5'
        
        # 
        result = r.text.splitlines()
        
        # 114
        char_index = result[114].rfind("")
        temp = result[114][:char_index]
        
        # temp100cnames
        cnames.append(temp)
    print(type(cnames[0]))
    names = cnames[0].split('')
    return names[:num]


names = getNames(50)  
```

```python
import requests
def getNames():
    times = 3
    cnames = []
    for x in range(times):
        url = 'http://www.richyli.com/name/index.asp'
        r = requests.get(url)
        r.encoding = 'big5'
        
        # 
        result = r.text.splitlines()
        
        # 114
        char_index = result[114].rfind("")
        temp = result[114][:char_index]
        
        # temp100cnames
        cnames.append(temp)    
    return cnames


names = getNames()
```

```python
#"",3
names_str = "".join(names)
names_list = names_str.split("") #299list
newLine_name = "\n".join(names_list)  #
print(newLine_name)
```
